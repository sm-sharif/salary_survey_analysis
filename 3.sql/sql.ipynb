```python
from sqlalchemy import create_engine

df = pd.read_csv('project_cleaned_v2.0.csv')

engine = create_engine('postgresql://postgres:sushmoy123@localhost:5432/project')

df.to_sql('salary_data', engine, if_exists='replace', index=False)

print(f"Rows: {len(df):,}")
```

```python
query1 = """
select 
job_category,
avg(annual_salary_usd) as avg_salary,
count(*) as total_employee
from salary_data
group by job_category
order by avg_salary desc
limit 10
"""

result1 = pd.read_sql(query1, engine)
print(result1)
```
